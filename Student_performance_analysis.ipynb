# Explorative Data analysis

# 1. Distribution of Average Scores of students

This visualization shows the distribution of students’ average scores. The goal is to understand how student performance is spread across different score ranges.
The choice of histogram because it helps to group continuous numerical data into intervals (bins), making it easier to observe patterns such as concentration, spread, and skewness in the dataset.

```python
plt.hist(df['Average_Score'])
plt.title("Distribution of Average Scores")
plt.xlabel("Average Score")
plt.ylabel("Number of Students")
plt.show()
```

### Insights and Interpretation
From the histogram of average student scores, several key observations can be made:
1.The majority of students score between **75 and 90**, indicating a generally strong academic performance across the dataset.
2.The distribution appears slightly **left-skewed**, suggesting that lower number of students fall into the lower score ranges i.e below 70.
3.Very few students score below **65**, which may indicate that underperformance is limited in this group.
4.There is a noticeable concentration of students who scored betweeb **80–90 range**, showing that this is the most common performance band.

### conclusion
The dataset suggests that student performance is generally high, with most students clustered in the upper-middle score range. This may reflect effective teaching, strong student engagement, or a selective dataset.

# 2. Study Hours vs Academic Performance

This visualization explores the relationship between students’ study hours and their average academic performance.

The objective is to determine whether increased study time leads to improved academic scores across subjects.

```python
plt.scatter(df['Study_Hours'], df['Average_Score'])
plt.xlabel("Study Hours")
plt.ylabel("Average Score")
plt.title("Study Hours vs Performance")
plt.show()
```

## Insights from Study Hours vs Performance

The scatter plot shows a general relationship between study hours and student performance.

### Key Observations:

-Students who study more hours (3–5 hours) tend to achieve higher average scores compared to those with fewer study hours (1–2 hours).
-There is a visible upward trend, suggesting a **positive relationship between study time and academic performance**.
-However, the relationship is not perfectly linear — some students with moderate study hours still perform at both high and average levels.
-This indicates that while study time is important, other factors (such as attendance, internet access, or extra tutorials) may also influence performance.

### Conclusion:
Overall, increased study hours are associated with better academic outcomes, but they may not be sole determinant of performance.

# 3. Attendance vs Performance
To analyze the relationship between student attendance and academic performance.
A boxplot is used to compare the distribution of scores between students with and without regular attendance.
This allows us to observe differences in median performance, variability, and potential outliers.

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8,5))
ax = sns.boxplot(x='Attendance_Rate', y='Average_Score', data=df)
plt.title("Impact of Attendance on Student Performance", fontsize=14, weight='bold')
plt.xlabel("Attendance")
plt.ylabel("Average Score")
plt.show()
```

### Interpretation

The box plot compares the distribution of average scores between students with regular attendance (**Yes**) and those without regular attendance (**No**).
-Students with regular attendance have a higher median score (approximately **84**) compared to students with irregular attendance (approximately **81**).
-The middle 50% of scores (Interquartile Range, IQR) for students who attended regularly ranges from about **77 to 90**, while for those with irregular attendance it ranges from about **73 to 88**.
-Both groups achieved similar maximum scores (around **95**), indicating that high-performing students exist in both attendance categories.
-The minimum score for regularly attending students is slightly higher (about **63**) than for irregular attendees (about **58**), suggesting fewer very low-performing students among those who attend regularly.
-No significant outliers are observed in either group, indicating relatively stable score distributions.

### Key Insights

-Regular attendance is associated with better academic performance, as evidenced by the higher median score.
-Students who attend classes consistently tend to achieve higher overall scores and experience fewer extremely low outcomes.
-The score distribution for regular attendees is shifted upward, suggesting a positive relationship between attendance and academic achievement.
-While attendance appears to improve performance, the overlap between the two distributions indicates that other factors may also influence student success.

###  Conclusion
Students with regular attendance generally achieve higher average scores and exhibit fewer low-performance cases, highlighting the positive impact of attendance on academic performance.

# 6.Performance based on schools
This analysis examines the average academic performance of students across different schools. By comparing average scores among schools, the study aims to identify performance variations and across the six schools that participated in the study.

```python
# Perfomance by School
school_perf = df.groupby('School')['Average_Score'].mean().sort_values()
plt.figure(figsize=(10,6))
ax = school_perf.plot(kind='barh')
for i, value in enumerate(school_perf):
    ax.text(value + 0.5, i, f"{value:.2f}", va='center', fontsize=10)
plt.title("Average Student Performance by School", fontsize=14, weight='bold')
plt.xlabel("Average Score")
plt.ylabel("School")
plt.grid(axis='x', linestyle='--', alpha=0.7)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.show()
```

### Interpretation of the Graph

The chart shows the average student performance across six schools. Overall, performance levels are relatively high, with all schools scoring between **~79 and ~85**, indicating generally strong academic outcomes across the board.
-Lagos Model Collegehas the highest average score (**85.19**), making it the top-performing school.
-Sunrise Secondary School (**83.00**) and Bright Future Academy (**82.19**) follow closely, forming a strong upper tier.
-Abesan High School(**81.58**) and Royal Scholars School (**81.15**) fall within the mid-performance range.
-Gr eenfield High Schoolhas the lowest average (**79.62**), though still within a competitive band.

### Key Insights

1.Narrow Performance Gap: The difference between the highest (**85.19**) and lowest (**79.62**) scores is relatively small (~5.5 points). This suggests minimal disparity in academic performance across schools.
2.Slight Lead by Top School-While Lagos Model College leads, the margin is not large, indicating that other schools are closely competitive.
3.Strong Middle Cluster- Most schools are concentrated between **81 and 83**, showing probably consistent teaching quality,similar academic standards and balanced student outcomes  
4.Opportunity for Improvement- Greenfield High School, despite being the lowest, is only slightly behind. Small improvements could significantly enhance its ranking.
5.Competitive Academic Environment
The close range of scores suggests a healthy competitive environment, which can encourage overall system-wide improvement.

### Summary
The graph highlights a well-balanced and high-performing educational system Although Lagos Model College ranks highest, the difference among schools is minimal, indicating consistency in academic delivery. Most schools perform within a tight range, reflecting uniform quality of education and strong competition, with opportunities for lower-ranked schools to quickly improve.

# Average Student Performance by Gender
This analysis explored the relationship between gender and academic performance by comparing the average scores of male and female students. The idea is that performance differences across gender groups can provide valuable insights into learning outcomes and help identify whether academic achievement is evenly distributed among students.

```python
gender_perf = (
    df.groupby('Gender')['Average_Score']
    .mean()
    .sort_values(ascending=False)
)

ax = gender_perf.plot(
    kind='bar',
    figsize=(8, 5)
)
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f')

plt.title('Average Student Performance by Gender', fontsize=16, fontweight='bold')
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Average Score', fontsize=12)

plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

### Key Insights

-The average scores for male and female students are very similar.
-Female students have a slightly higher average score compared to male students.
-The difference between the two groups is minimal and may not be statistically significant.

### Interpretation

This suggests that gender does not show any significant difference in academic performance between male and female students. While female students show a slightly higher average score, the gap is minimal and does not indicate a meaningful disparity.

### Conclusion

The result suggests that gender is not a strong determinant of student performance within this dataset. Academic outcomes appear to be influenced more by other factors such as study habits, attendance, and access to learning resources.
Overall, the results indicate a relatively balanced performance distribution across genders, highlighting fairness in outcomes and suggesting that interventions to improve performance should focus on behavioral and environmental factors rather than gender.

4.Positive Overall Academic Performance-The concentration of students in the High Performance category indicates a generally strong academic profile across the dataset.

### Interpretation

The performance distribution reveals that academic achievement is heavily skewed toward the High Performance category. ### Conclusion

With approximately **89%** of students classified as High Performers and no students categorized as Low Performers, the dataset reflects strong educational outcomes. # Correlation Analysis of Academic Performance Factors
This explores the relationships among key academic performance indicators, including subject scores, study hours, and attendance rate. A correlation matrix was used to measure the strength and direction of relationships between variables, helping to identify patterns and factors associated with student achievement.

```python
df['Attendance_Rate'] = df['Attendance_Rate'].map({
    'Yes': 1,
    'No': 0
})
import seaborn as sns

sns.heatmap(df[['Math_Score','English_Score','Science_Score','Study_Hours','Attendance_Rate']].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

 ### Key Insights

1.Strong Positive Correlation Among Subject Scores-The correlation matrix reveals a very strong positive relationship among the three academic subjects. English Score and Science Score exhibit the strongest correlation (**0.99**), while Math Score shows strong correlations with both English Score (**0.92**) and Science Score (**0.93**). This suggests that students who perform well in one subject tend to perform well in others.
2.Study Hours Have a Weak Positive Relationship with Academic Performance-Study Hours showed weak positive correlations with Math Score (**0.13**), English Score (**0.23**), and Science Score (**0.20**). Although students who study more tend to achieve slightly better scores, the relationship is not particularly strong in this dataset.
3.Academic Performance Appears Consistent Across Subjects-The high correlations among subject scores indicate consistency in student achievement. High-performing students generally perform well across all subjects rather than excelling in only one area.

### Interpretation
The correlation analysis suggests that academic performance is highly interconnected across subjects. Students who excel in one subject are very likely to achieve strong results in other subjects as well. This pattern may indicate that underlying factors such as learning ability, discipline, motivation, and access to educational resources contribute broadly to student achievement.
The relatively weak relationship between Study Hours and academic scores implies that simply increasing study time may not guarantee significantly better performance. The effectiveness of study habits and other supporting factors may be more important than the number of hours spent studying.
It is also important to note that correlation does not imply causation. While variables may move together, the analysis does not prove that one variable directly causes changes in another.

### Conclusion

Overall, the correlation matrix highlights strong relationships among academic subjects and a relatively weak relationship between Study Hours and performance. The findings suggest that student success is influenced by a combination of factors rather than study time alone. Future analyses should explore additional variables such as attendance, internet access, school environment, and extracurricular involvement to gain a more comprehensive understanding of the drivers of academic performance.
